```
%%time
%%bash
set -e
ENV_FILE=/mydrive/language-modeling/environment.yml
ENV_NAME=$(head -n 1 $ENV_FILE | cut -d ':' -f 2 | cut -d ' ' -f 2)

FILE_NAME=$ENV_NAME-env.tar.gz
ENVS_PATH=/mydrive/envs
FILE=$ENVS_PATH/$FILE_NAME
if [ -f $FILE ]; then
    echo "found $FILE"
    cp $FILE ./
    tar xvzf $FILE_NAME -C /usr/local/
else 
  echo "/usr/local/envs/$ENV_NAME"
  if [ -d "/usr/local/envs/$ENV_NAME" ]; then
    echo "updating env $ENV_NAME"
    conda env update -f $ENV_FILE
  else
    echo "creating env from $ENV_FILE"
    conda env create -f $ENV_FILE
  fi
  tar -czvf $FILE_NAME /usr/local/envs
  cp $FILE_NAME $ENVS_PATH/
fi
```

```
%%time
%%bash
source activate language_modeling
# locally do: python write_wandb_key.py
export WANDB_API_KEY=$(head -n 1 /mydrive/wandb.key | cut -d '=' -f 2 )
export WANDB_PROJECT=language-modeling
python -m wandb init --project language-modeling

export TRAIN_FILE=/content/wikitext-2-raw/wiki.train.raw
export TEST_FILE=/content/wikitext-2-raw/wiki.test.raw
ls -althr $TRAIN_FILE

cd transformers/examples/language-modeling/
rm -rf output
python run_language_modeling.py \
    --output_dir=output \
    --run_name=distilbert-2-epoch-eval-bs16 \
    --logging_steps=50 \
    --num_train_epochs=2 \
    --model_type=bert \
    --model_name_or_path=distilbert-base-uncased \
    --do_train \
    --do_eval \
    --mlm \
    --train_data_file=$TRAIN_FILE \
    --eval_data_file=$TEST_FILE \
    --per_gpu_train_batch_size 16 --per_gpu_eval_batch_size 16
```
